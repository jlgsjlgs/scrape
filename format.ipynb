```python
import pandas as pd
import re
```

```python
df = pd.read_excel('./excel/royal.xlsx')
df.insert(2,'FormattedDate', None)
df
```

```python
# Increase the number of rows and columns to display
pd.set_option('display.max_rows', None)    # Display all rows
pd.set_option('display.max_columns', None)  # Display all columns

month2num = {'January':1,
            'February':2,
            'March':3,
            'April':4,
            'May':5,
            'June':6,
            'July':7}


for idx in df.index:
    date = df['Date'][idx][:-4]
    
    # Use regular expressions to extract the day and month from the input string
    match = re.match(r'(\d+)([A-Za-z]+)', date)
    if match:
        day = match.group(1)
        month_name = match.group(2)

        # Convert the month name to its numeric equivalent
        month = month2num.get(month_name)

        if month is not None:
            # Format the day and month as MM-DD
            if len(day) == 1:
                day = "0" + day
            formatted_date = f"2023-{month:02}-{day}"
            df['FormattedDate'][idx] = formatted_date

df
```

```python
data_categories = {
    "Intellectual Property", # e.g. Patents
    "Financial Data", # Customer/Employee financial info or company info, payroll
    "Customer Data", # e.g. Clients full info (anyth related to JUS CLIENT/CUSTOMERS)
    "Personal Data", # e.g. Phone number and email address, ID, passport (INDIV    LEVEL)
    "Employee Data", # position, work number
    "Sensitive Communications", # Emails, leaked convos
    "Strategic Plans",   # Business deals, acquisitions
    "Healthcare Records", 
    "Classified Information", # Company data, client and employee documents, payroll
    "Research and Development Data",
    "Legal and Compliance Information" # Contracts, NDAs
}


for categories in data_categories:
    df[categories] = 0

for idx in df.index:
    stolen = df["Stolen"][idx]

    if isinstance(stolen, str):
        stolen_split = stolen.split(",")
        for items in stolen_split:
            if items[0] == " ":
                items = items[1:]
            
            if items in data_categories:
                df[items][idx] = 1

df  
```

```python
df.to_excel('output.xlsx')
```
